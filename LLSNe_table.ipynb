# Making the table of previous LL SNe II for the paper

```python
import pandas as pd
import os
```

```python
df = pd.read_csv(os.getcwd()+'/LLSNe/LL_expds_hand.csv')
df
```

```python
for sn in df.name:
    row = df[(df['name']==sn)]
    host = row['host'].values[0]
    print('SN '+sn+' & '+str(row['expd'].values[0])+' ('+str(row['expderr'].values[0])+') & '+\
          "{:.3f}".format(round(row['z'].values[0],3))+' & '+"{:.2f}".format(row['mu'].values[0])+' & '+\
          host+' & \citet{'+','.join(row['ref'].values[0].split('/'))+r'} \\')
```
